```python
import pandas as pd
df = pd.read_csv(r'D:\US-election clasand reg\data\US.csv')

df
```

```python
# Define the columns with percentages
percentage_columns = [
    '2020 Democrat vote %',
    '2020 Republican vote %',
    '2020 other vote %'
  
    # Add any other relevant percentage columns
]

# Convert percentage columns to numeric after removing '%' symbol
for col in percentage_columns:
    # Check if the column's dtype is object (string), and apply the conversion if true
    if df[col].dtype == 'object':
        df[col] = df[col].str.rstrip('%').astype('float') / 100
```

```python
# Create 'Population density' by dividing 'Total Population' by 'Area in square Km'
df['Population density'] = df['Total Population'] / df['Area in square Km']

# Verify the creation of 'Population density'
print(df[['Total Population', 'Area in square Km', 'Population density']].head())
# Rename columns if needed for easier access
df.rename(columns={'Unnamed: 0': 'Index'}, inplace=True)
```

```python
df['Winner'] = df.apply(lambda row: 'Republican' if row['2020 Republican vote %'] > row['2020 Democrat vote %'] else 'Democratic', axis=1)
df['Winner_binary'] = df['Winner'].map({'Republican': 1, 'Democratic': 0})
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Apply LabelEncoder to the 'state' column
label_encoder = LabelEncoder()
df['state_encoded'] = label_encoder.fit_transform(df['state'])

# Step 2: Drop the 'state', 'county', and 'index' columns
df = df.drop(columns=['state', 'county', 'Index'])

# Step 3: Display the state to encoded label mapping
state_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("State Encoding Mapping:")
for state, encoded_label in state_mapping.items():
    print(f"{state} - {encoded_label}")
```

```python
import pandas as pd
import numpy as np

# Define columns for income that contain commas (we'll clean them and convert to numeric)
income_columns = ['Median income (dollars)', 'Mean income (dollars)']

# Convert income columns (ensure they are strings first, then remove commas and convert to numeric)
for col in income_columns:
    # Replace commas and handle invalid values (e.g., '-')
    df[col] = df[col].replace({'-': np.nan, ',': ''}, regex=True)  # Replace '-' with NaN and commas
    df[col] = df[col].astype(str).astype(float)  # Convert to float after cleaning

# Now check the data types
print(df.dtypes)
```

```python
import pandas as pd

# Define columns that contain percentage values (we'll clean them and convert to numeric)
percentage_columns = [
    'Population with less than 9th grade education', 
    'Population with 9th to 12th grade education, no diploma',
    'High School graduate and equivalent', 
    'Some College,No Degree',
    'Associates Degree', 
    'Bachelors Degree', 
    'Graduate or professional degree'
]

# Convert percentage columns (ensure they are strings first, then remove '%' and convert to float)
for col in percentage_columns:
    df[col] = df[col].astype(str).str.replace('%', '').astype(float)

# Define columns for income that contain commas (we'll clean them and convert to numeric)
income_columns = ['Median income (dollars)', 'Mean income (dollars)']

# Convert income columns (ensure they are strings first, then remove commas and convert to numeric)
for col in income_columns:
    df[col] = df[col].astype(str).str.replace(',', '').astype(float)

# Now check the data types
print(df.dtypes)
```

```python
import pandas as pd
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split


# Select only the 3 education-related features
X = df[['Population with less than 9th grade education', 
        'Population with 9th to 12th grade education, no diploma', 
        'High School graduate and equivalent']]  # 3 most relevant education features

# Define individual targets (assuming these columns exist in your DataFrame)
y_democrat_vote_pct = df['2020 Democrat vote %']
y_republican_vote_pct = df['2020 Republican vote %']
y_democrat_vote_raw = df['2020 Democrat vote raw']
y_republican_vote_raw = df['2020 Republican vote raw']

# Split the data for each target
X_train, X_test, y_train_dem, y_test_dem = train_test_split(X, y_democrat_vote_pct, test_size=0.3, random_state=42)
_, _, y_train_rep, y_test_rep = train_test_split(X, y_republican_vote_pct, test_size=0.3, random_state=42)
_, _, y_train_dem_raw, y_test_dem_raw = train_test_split(X, y_democrat_vote_raw, test_size=0.3, random_state=42)
_, _, y_train_rep_raw, y_test_rep_raw = train_test_split(X, y_republican_vote_raw, test_size=0.3, random_state=42)

# Initialize and fit individual regressors for each target
reg_dem_pct = HistGradientBoostingRegressor(random_state=42)
reg_dem_pct.fit(X_train, y_train_dem)

reg_rep_pct = HistGradientBoostingRegressor(random_state=42)
reg_rep_pct.fit(X_train, y_train_rep)

reg_dem_raw = HistGradientBoostingRegressor(random_state=42)
reg_dem_raw.fit(X_train, y_train_dem_raw)

reg_rep_raw = HistGradientBoostingRegressor(random_state=42)
reg_rep_raw.fit(X_train, y_train_rep_raw)

# Make predictions for each target and calculate Mean Squared Error (MSE)
y_pred_dem_pct = reg_dem_pct.predict(X_test)
y_pred_rep_pct = reg_rep_pct.predict(X_test)
y_pred_dem_raw = reg_dem_raw.predict(X_test)
y_pred_rep_raw = reg_rep_raw.predict(X_test)

mse_dem_pct = mean_squared_error(y_test_dem, y_pred_dem_pct)
mse_rep_pct = mean_squared_error(y_test_rep, y_pred_rep_pct)
mse_dem_raw = mean_squared_error(y_test_dem_raw, y_pred_dem_raw)
mse_rep_raw = mean_squared_error(y_test_rep_raw, y_pred_rep_raw)

# Print MSE results
print(f"Regression MSE for Democrat Vote %: {mse_dem_pct:.4f}")
print(f"Regression MSE for Republican Vote %: {mse_rep_pct:.4f}")
print(f"Regression MSE for Democrat Vote Raw: {mse_dem_raw:.4f}")
print(f"Regression MSE for Republican Vote Raw: {mse_rep_raw:.4f}")
```
